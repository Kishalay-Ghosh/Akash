```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/Salary_Data.csv")
print(data)
```

```python
sns.heatmap(data.isnull()) #find the null(not field data) return 1 when null
```

```python
avgsal=data['Salary'].mean()#for mean data of a perticuler column
data['Salary'].fillna(avgsal,inplace=True) #fill avaerage value in place of null
```

```python
sns.heatmap(data.isnull()) #find the null(not field data) return 1 when null
```

```python
x=data.iloc[:,:-1].values #[:]--> row and column ,-1 means delete one row from end
print(x)
y=data.iloc[:,1].values
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
#print(x_train.shape)


```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
inp=[[2.5]]   #input for model
y_pred=model.predict(inp)  #model is predect
print(y_pred)

#plot
plt.scatter(inp,y_pred,color="red")
plt.scatter(x_train,y_train,color="green")
plt.xlabel("Experience")
plt.ylabel("Salary")
```
